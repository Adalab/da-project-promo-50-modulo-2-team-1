```python
#Last.fm
API_KEY = 'f0eb46d5c17e5b9b37b20631a366bb20'  
BASE_URL = 'http://ws.audioscrobbler.com/2.0/'
URL_BIOGRAFIA=f"http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&artist=artist&api_key={API_KEY}&format=json"
```

```python
info_artista=[]
for artista in lista_artistas:
    URL_BIOGRAFIA=f"http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&artist={artista}&api_key={API_KEY}&format=json"
    params_info = {
        'method': 'artist.getinfo',
        'artist': artista,
        'api_key': API_KEY,
        'format': 'json'
        }
    response_info = requests.get(URL_BIOGRAFIA, params=params_info)
    data_info = response_info.json()
    
    artist_data = {
        'Artist': artista,
        'Biography': data_info['artist'].get('bio', {}).get('summary'),
        'Listeners': data_info['artist'].get('stats', {}).get('listeners', "Sin datos"),
        'Playcount': data_info['artist'].get('stats', {}).get('playcount', "Sin datos")
        }
    info_artista.append(artist_data)
info_artista
```

```python
URL_SIMILAR="http://ws.audioscrobbler.com/2.0/?method=artist.getsimilar&artist=cher&api_key=YOUR_API_KEY&format=json"
```

```python
artista_similar = []

for artista in lista_artistas:
    URL_SIMILAR = f"http://ws.audioscrobbler.com/2.0/?method=artist.getsimilar&artist={artista}&api_key={API_KEY}&format=json"
    params_info = {
        'method': 'artist.getsimilar',
        'artist': artista,
        'api_key': API_KEY,
        'format': 'json'
    }
    response_info = requests.get(URL_SIMILAR, params=params_info)
    data_info = response_info.json()
    
    similares = data_info['similarartists'].get('artist', [])
    
    # Creamos una lista con los nombres de varios artistas similares
    nombres_similares = [artista_similar['name'] for artista_similar in similares]
    
    artist_data = {
        'Artist': artista,
        'Similar': nombres_similares  # guardamos la lista completa
    }
    artista_similar.append(artist_data)

print(artista_similar)
```

```python
URL_top_canciones=f'http://ws.audioscrobbler.com/2.0/?method=artist.gettoptracks&artist={artista}&api_key={API_KEY}&format=json'
```

```python
top_canciones=[]
for artista in lista_artistas:
    URL_top_canciones = f"http://ws.audioscrobbler.com/2.0/?method=artist.gettoptracks&artist={artista}&api_key={API_KEY}&format=json"
    
    response_info = requests.get(URL_top_canciones)
    data_info = response_info.json()

    # Obtener la lista de canciones
    canciones_top = []
    
    if "toptracks" in data_info and "track" in data_info["toptracks"]:
        for track in data_info["toptracks"]["track"]:
            artist_data = {
                'Artist': artista,
                'Track': track.get('name', "Sin nombre"),
                'Listeners': track.get('listeners', "Sin datos"),
                'Playcount': track.get('playcount', "Sin datos")
            }
            canciones_top.append(artist_data)

    top_canciones.extend(canciones_top)
top_canciones
```
