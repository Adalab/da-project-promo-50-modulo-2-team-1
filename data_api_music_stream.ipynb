4. Spotify: recopilación canciones años 2019-2024 para rock, pop, reggaeton y rap

```python
def tracks_genre (sp, genre, limit=50, total_limit=1000): # Función para obtener 1000 canciones de un género determinado 
    tracks_genre= []
    for offset in range (0, total_limit, limit): # Iteramos por las distintas páginas hasta colectar 1000 canciones
        query = f'genre:{genre} year:2019-2024'
        data = sp.search(q= query, type='track', limit=limit, offset=offset)

        for item in data['tracks']['items']: # Iteramos en la lista que nos da el API y sacamos los datos que nos interesan para añadirlos a la lista tracks_genre
            tracks_genre.append({
                    'name_track': item['name'],
                    'name_artist': item['artists'][0]['name'],
                    'genre': genre,
                    'type': item['type'],
                    'year': item['album']['release_date'][:4] })
            
    df_tracks_genre= pd.DataFrame(tracks_genre)
    df_tracks_genre.to_csv(f'{genre}_tracks.csv') # Guardamos las 1000 canciones en forma de csv

    print(df_tracks_genre.head())

    return f'{genre}_tracks.csv'
```

5. Spotify: recopilación albums años 2019-2024 para rock, pop, reggaeton y rap

7. Last.fm: obtener biografía, listeners y playcounts de los artistas encontrados en Spotify

```python
artist_info=[] # Lista para guardar la información de los artistas

for artist in all_artist:
    URL_BIO=f"http://ws.audioscrobbler.com/2.0/?method=artist.getinfo&artist={artist}&api_key={API_KEY}&format=json" # Entramos al API cambiando el nombre del artista en la URL
    params_info = {
        'method': 'artist.getinfo',
        'artist': artist,
        'api_key': API_KEY,
        'format': 'json'
        }
    response_info = requests.get(URL_BIO, params=params_info)
    data_info = response_info.json()
    
    if 'artist' in data_info:
        artist_data = {
            'Artist': artist,
            'Biography': data_info['artist'].get('bio', {}).get('summary'),
            'Listeners': data_info['artist'].get('stats', {}).get('listeners', "Sin datos"),
            'Playcount': data_info['artist'].get('stats', {}).get('playcount', "Sin datos")
            }
        artist_info.append(artist_data)
    else: # Si el artista no aparece en el API nos informa de esta forma
        artist_data = {
            'Artist': artist,
            'Biography': 'None',
            'Listeners': 'None',
            'Playcount': 'None'
            }
        artist_info.append(artist_data)

df_artist_info= pd.DataFrame(artist_info) 

df_artist_info.to_csv('artist_info.csv') # Guardamos la información del artista en csv

df_artist_info.head()
```

```python
similar_artist=[] # Lista para guardar los artistas similares

for artist in all_artist:
    URL_SIMILAR = f"http://ws.audioscrobbler.com/2.0/?method=artist.getsimilar&artist={artist}&api_key={API_KEY}&format=json" # Entramos al API cambiando el nombre del artista en la URL    params_info = {
    params_info = {
        'method': 'artist.getsimilar',
        'artist': artist,
        'api_key': API_KEY,
        'format': 'json'
    }
    response_info = requests.get(URL_SIMILAR, params=params_info)
    data_info = response_info.json()

    similar = data_info.get('similarartists', {}).get('artist', [])
    
    name_similar = [similar_artist['name'] for similar_artist in similar] # Creamos una lista con los nombres de varios artistas similares
    
    if similar:
        name_similar = [similar_artist['name'] for similar_artist in similar]

        artist_data = {
            'Artist': artist,
            'Similar': name_similar
        }
    else: # Si el artista no tiene esta info en la API aparece de esta forma
        artist_data = {
            'Artist': artist,
            'Similar': None
        }
    
    similar_artist.append(artist_data)

df_similar_artist= pd.DataFrame(similar_artist) 

df_similar_artist.to_csv('similar_artist.csv') # Guardamos la información del artista en csv

df_similar_artist.head()
```

```python
top_tracks = []  # Lista para guardar la información de las canciones top

for artist in all_artist:
    URL_TOP_TRACKS = f"http://ws.audioscrobbler.com/2.0/?method=artist.gettoptracks&artist={artist}&api_key={API_KEY}&format=json"
    
    response_info = requests.get(URL_TOP_TRACKS)
    data_info = response_info.json()

    if "toptracks" in data_info and "track" in data_info["toptracks"]:
        for track in data_info["toptracks"]["track"]:
            artist_data = {
                'Artist': artist,
                'Track': track.get('name', None),
                'Listeners': track.get('listeners', None),
                'Playcount': track.get('playcount', None)
            }
            top_tracks.append(artist_data)
    else: # Si no hay info del artista, añadimos una entrada con valores None
        artist_data = {
            'Artist': artist,
            'Track': None,
            'Listeners': None,
            'Playcount': None
        }
        top_tracks.append(artist_data)
                                       
df_top_tracks = pd.DataFrame(top_tracks)
df_top_tracks.to_csv('top_tracks.csv')  # Guardamos en csv

df_top_tracks.head()
```
